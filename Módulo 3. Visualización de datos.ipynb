## Anatomía de un gráfico con Matplotlib

```python
# Ejemplo: creación de una figura
import matplotlib.pyplot as plt

# Create a figure with a specific size
fig = plt.figure(figsize=(10, 6))  # Width: 10 inches, Height: 6 inches
print(fig)
```

```python
# Ejemplo: creación de subparcelas
import matplotlib.pyplot as plt

# Create a figure with 2 rows and 2 columns of subplots
fig, ax = plt.subplots(2, 2)

# Access individual subplots using indexing
ax[0, 0].plot([1, 2, 3], [4, 5, 6])  # Top-left subplot
ax[0, 1].bar(['A', 'B', 'C'], [7, 8, 9])  # Top-right subplot
ax[1, 0].scatter([10, 20, 30], [11, 12, 13])  # Bottom-left subplot
ax[1, 1].hist([1, 1, 2, 3, 3, 3])  # Bottom-right subplot

plt.show()
```

```python
# Ejemplo: añadir etiquetas y título
import matplotlib.pyplot as plt

x = [1, 2, 3]
y = [4, 5, 6]

plt.plot(x, y)
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Plot Title')
plt.show()
```

```python
# Ejemplo: Personalización de marcas y etiquetas de marcas
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100) # Generate 100 evenly spaced points from 0 to 10 (y los une con el seno de x)
y = np.sin(x)

plt.plot(x, y)

# Set tick locations and labels for the x-axis (etiquetas de marcas en el eje x)
plt.xticks(np.arange(0, 11, 2), ['0', '2', '4', '6', '8', '10'])

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

plt.show()
```

```python
# Ejemplo: Añadir una leyenda
import matplotlib.pyplot as plt

x = [1, 2, 3]
y1 = [4, 5, 6]
y2 = [7, 8, 9]

plt.plot(x, y1, label='Data Series 1')
plt.plot(x, y2, label='Data Series 2')

plt.legend()  # Automatically creates a legend based on labels (Data series 1 y data series 2)
plt.show()
```

## Galería Matplotlib

### Gráficos de líneas

```python
# Ejemplo de gráfico de líneas sencillo

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)  # Generate 100 evenly spaced points from 0 to 10 (y los une con el seno de x)
y = np.sin(x)  # Compute the sine of each x value

plt.plot(x, y)
plt.xlabel('Time (days)')  # Label the x-axis
plt.ylabel('Stock Price ($)')  # Label the y-axis
plt.title('Stock Price Trend')  # Add a title
plt.show()  # Display the plot
```

- Ventaja: Conecta puntos para guiar la vista a lo largo de la trayectoria de los datos, resaltando flujos, reflujos y tendencias.

### Gráficos de dispersión

```python
# Ejemplo de gráfico de dispersión

import matplotlib.pyplot as plt
import numpy as np

ages = np.linspace(20, 60, 100)  # Ages ranging from 20 to 60 (100 edades de entre 20 y 60 años)
income = ages * 500 + np.random.normal(0, 10000, 100)  # Simulating a positive correlation with some noise

plt.scatter(ages, income)
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Age vs. Income')
plt.show()
```

- Ventaja: Muestra densidad, dispersión, valores atípicos y posibles correlaciones, enfatizando los puntos individuales sin líneas de conexión.

### Gráficos de barras

```python
# Ejemplo de gráfico de barras

import matplotlib.pyplot as plt

categories = ['Product A', 'Product B', 'Product C', 'Product D']
values = [1500, 2300, 1200, 3000]

plt.bar(categories, values)
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Product Sales Comparison')
plt.show()
```

- Ventaja: Claridad y simplicidad; facilita la comparación de magnitudes (comparar valores entre distintas categorías), identificación de patrones y detección de valores atípicos.


### Histogramas

```python
# Ejemplo de histograma

import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)  # Generate 1000 random numbers from a standard normal distribution (tiene forma de la distribución normal)

plt.hist(data, bins=30)  # Create a histogram with 30 bins (picos)
plt.xlabel('Exam Score')
plt.ylabel('Number of Students')
plt.title('Distribution of Exam Scores')
plt.show()
```

- Uso: Mostrar la distribución de una variable numérica y analizar su forma, tendencia central, dispersión y posibles valores atípicos.
- Ventaja: Permite identificar rangos comunes de valores, asimetrías, valores atípicos y características generales de los datos.

### Gráficos de sectores o circulares (proporciones y porcentajes)

```python
# Ejemplo de gráfico de sectores o circular (porcentajes)

import matplotlib.pyplot as plt

labels = ['Windows', 'macOS', 'Linux', 'Other']
sizes = [70, 20, 5, 5]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')  # Display percentages on each slice
plt.title('Market Share of Operating Systems')
plt.show()
```

- Ventaja: Aprovecha la percepción humana de tamaños relativos para comunicar de manera clara la contribución de cada categoría.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
music_data = pd.read_csv('music_data.csv')

# Display the first 5 rows
print("First 5 rows of the dataset:")
print(music_data.head())
```

## Step 3: Generating a bar chart using Matplotlib's pyplot module

Follow the steps in the instructions and run the cell below.

```python
# Count the occurrences of each unique value in the `genre` column
genre_counts = music_data['genre'].value_counts()   

# Sort the genres by frequency in descending order and select the top 10 genres
top_10_genres = genre_counts.head(10)

# Create a bar chart to display the top 10 genres by frequency
plt.figure(figsize=(10, 6)) # dimensiones del gráfico
top_10_genres.plot(kind='bar')

# Add title and labels
plt.title('Top 10 Genres by Frequency') # insert code here 
plt.xlabel('Genre') # insert code here 
plt.ylabel('Number of Songs') # insert code here 

# Display the plot using plt.show()
plt.show()
```

## Step 4: Generating a histogram using Matplotlib's plt.figure()

Follow the steps in the instructions and run the cell below.

```python
# Extract the duration_seconds column
duration_seconds = music_data['duration_seconds'] # insert code here 

# Create a histogram to display the distribution of `duration_seconds`
plt.figure(figsize = (8, 5)) # insert code here 
plt.hist(duration_seconds, bins=20, color='skyblue', edgecolor='black')

# Add title and labels
plt.title('Distribution of Song Durations')
plt.xlabel('Duration (seconds)')
plt.ylabel('Frequency')

# Display the plot using plt.show()
plt.show()
```

## Step 5: Generating a histogram using Matplotlib's plt.hist()

Follow the steps in the instructions and run the cell below.

```python
# Extract the popularity column
popularity = music_data['popularity'] # insert code here 

# Create a histogram to display the distribution of popularity 
plt.figure(figsize=(8, 5))  # Set the figure size for better readability

# Plot the histogram with customized bins and colors
plt.hist(popularity, bins=10, color='green', edgecolor='black')  # insert code here

# Add title and labels
plt.title('Distribution of Song Popularity')
plt.xlabel('Popularity')
plt.ylabel('Frequency')

# Display the plot using plt.show()
plt.show()
```

## Step 6: Generating heatmaps using Plotly

Follow the steps outlined in your instructions document to complete the code below. Click run to see the results.

```python
import plotly.express as px

# Create a new dataframe `numeric_features` containing only the 
# numeric columns we want to analyze
numeric_features = music_data[['tempo', 'energy', 'danceability', 
                               'valence', 'acousticness', 'popularity']]

# Compute the correlation matrix 
correlation_matrix = numeric_features.corr()

# Create and display a heatmap using Plotly (mapa de calor de plotly) canciones con mucha enería son muy bailables
plt = px.imshow(correlation_matrix,
                x=correlation_matrix.columns,
                y=correlation_matrix.columns,
                color_continuous_scale="RdBu",  # Use a built-in Plotly colorscale
                text_auto = ".2f",
                title="Correlation Matrix of Music Features"
               )
plt.show()
```

## Step 7: Generating violin plots using Plotly

Follow the steps outlined in your instructions document to complete the code below. Click run to see the results.

```python
import plotly.express as px

plt = px.violin(music_data, 
                x='genre', 
                y='popularity', 
                box = True , 
                title="Popularity Distribution Across Genres")

plt.show()
```
